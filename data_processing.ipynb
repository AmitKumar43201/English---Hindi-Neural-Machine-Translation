```python
from pathlib import Path

root = Path("/content/drive/MyDrive/MachineLearning")
```

```python
import re
from datasets import load_dataset
import unicodedata

dataset = load_dataset(
    "sentence-transformers/parallel-sentences-opensubtitles",
    "en-hi"
)

train = dataset["train"]
i = 0 


def normalize(text):
    return unicodedata.normalize("NFKC", text)

def clean_en(text):
    text = normalize(text)
    text = re.sub(r"[^a-zA-Z0-9 .,!?'\-]", "", text)
    return text.strip()

def clean_hi(text):
    text = normalize(text)
    text = re.sub(r"[^\u0900-\u097F0-9 .,!?'\-]", "", text)
    return text.strip()

def clean_line(text):
    text = text.strip()
    if text.startswith("-"):
        text = text.lstrip("- ").strip()
    return text

def is_mostly_numeric(text):
    digits = sum(c.isdigit() for c in text)
    return digits > len(text) * 0.4


with open(f"{root}/train.en", "w", encoding="utf-8") as f_en, \
     open(f"{root}/train.hi", "w", encoding="utf-8") as f_hi:

    for example in train:
        en = example["english"]
        hi = example["non_english"]
        
        en = normalize(en)
        hi = normalize(hi)

        en = clean_line(en)
        hi = clean_line(hi)

        if is_mostly_numeric(en) or is_mostly_numeric(hi):
            continue

        en = clean_en(en)
        hi = clean_hi(hi)

        en = " ".join(en.split())
        hi = " ".join(hi.split())

        if len(en.split()) < 2:
            continue

        if not en or not hi:
            continue

        if len(en.split()) > 3 and len(hi.split()) == 1:
            continue

        if len(en) > 200 or len(hi) > 200:
            continue
        
        f_en.write(en + "\n")
        f_hi.write(hi + "\n")
        i += 1
print(f"Total {i} rows downloaded")
```
